### Lets Handle Outliers


df['Purchase Amount (USD)']=df['Purchase Amount (USD)'].astype('float')

```python
Q1=df['Purchase Amount (USD)'].quantile(0.25)
Q3=df['Purchase Amount (USD)'].quantile(0.75)

IQR=Q3-Q1
lower_fence=IQR-1.5*(Q1)
high_fence=IQR+1.5*(Q3)
outliers=df[(df['Purchase Amount (USD)']<lower_fence) | (df['Purchase Amount (USD)']>high_fence)]


print(outliers.shape[0])
```

```python
print(df[df['Purchase Amount (USD)']>high_fence])
```

```python
# if(df['Purchase Amount (USD)']>high_fence):
df['Purchase Amount (USD)']=df['Purchase Amount (USD)'].clip(upper=high_fence)
```

```python
count = (df['Purchase Amount'] > high_fence).sum()
print("Count:", count)
```
